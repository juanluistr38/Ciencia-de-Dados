```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregando o arquivo CSV

df = pd.read_csv('/content/canadian_immegration_data.csv')

# Filtrando os dados para Argentina e Brasil
df_filtered = df[df['Country'].isin(['Argentina', 'Brazil'])]

# Selecionando as colunas que representam os anos
anos = list(map(str, range(1980, 2014)))

# Separando os dados por paÃ­s
argentina = df_filtered[df_filtered['Country'] == 'Argentina'][anos].values.flatten()
brasil = df_filtered[df_filtered['Country'] == 'Brazil'][anos].values.flatten()

# Criando o eixo x (anos)
x_values = list(map(int, anos))  # Convertendo os anos para inteiros para o eixo x

plt.figure(figsize=(20,10))

# Plotando os dados
plt.plot(x_values, argentina, marker='o', linestyle='-', color='blue', label='Argentina')
plt.plot(x_values, brasil, marker='o', linestyle='-', color='green', label='Brasil')

# Personalizando o grÃ¡fico
plt.xlabel('Ano')
plt.ylabel('NÃºmero de Imigrantes')
plt.title('ComparaÃ§Ã£o de Imigrantes entre Brasil e Argentina (1980-2013)')
plt.legend()

# mostrabdi os anos um a um
plt.xticks(x_values)


# Exibindo o grÃ¡fico
plt.show()
```
