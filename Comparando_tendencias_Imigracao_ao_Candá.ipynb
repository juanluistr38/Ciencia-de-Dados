```python
import pandas as pd
import matplotlib.pyplot as plt

# Armazenando o CSV
df_imigracao = pd.read_csv('/content/canadian_immegration_data.csv')

# Variável que que armezena lista de anos que servirá de variável independente no eixo x.
anos = list(map(str,range(1980,2014)))

#Para que seja possível filtrar por Brasil:
df_imigracao.set_index('Country',inplace=True)

# Variável que armazenará as colunas e linhas filtrando pelos rótulo Brazil e pela lista de rotulos anos
brasil = df_imigracao.loc['Brazil', anos]

# Variável que fornecerá uma estrutura de chave-valor, precursora para a formacao de DataFrame menor com 2 colunas
brasil_dict={'ano': brasil.index.tolist(),'imigrantes': brasil.values.tolist()}

# Variável que armazenará a tabela pronta apenas com valores do Brasil durante os anos
df_brasil = pd.DataFrame(brasil_dict)

# Variável que armazenará as colunas e linhas filtrando pelos rótulo Argentina e pela lista de rotulos anos
argentina = df_imigracao.loc["Argentina",anos]

# Variável que fornecerá uma estrutura de chave-valor, precursora para a formacao de DataFrame menor com 2 colunas
argentina_dict = {'ano':argentina.index.tolist(),"imigrantes": argentina.values.tolist()}

#Variável que armazenará a tabela pronta apenas com valores da Argentina durante os anos
df_argentina = pd.DataFrame(argentina_dict)

# Valores para o eixo x
x = df_argentina['ano']

# Valores da Argentina para o eixo y
y = df_argentina["imigrantes"]

# Valores do Brasil para o eixo y
y1 = df_brasil['imigrantes']


# Tamanho do gráfico de linha
plt.figure(figsize=(20,12))

# Grafico com a linha da Argentina
plt.plot(x,y,marker='o',linestyle='-',color="b",label='Imigrantes da Argentina')

# Grafico com a linha do Brasil
plt.plot(x,y1, marker ='s', linestyle='--',color ='green',label='Imigrantes do Brasil')


# Título do gráfico
plt.title("Comparação entre Imigrantes da Argentina e Brasil")

# Legenda
plt.xlabel("Anos",fontsize =12,color="black",fontweight="bold" )
plt.ylabel("Número de Imigrantes",fontsize =12,color="black",fontweight="bold")

# Marcações para o eixo x
x = [1980, 1981, 1982, 1983, 1984, 1985, 1986, 1987, 1988, 1989, 1990,
     1991, 1992, 1993, 1994, 1995, 1996, 1997, 1998, 1999, 2000,
     2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010]

#Marcações para o eixo y
y_ticks = [i*1 for i in range(1,2701,100) ]
plt.yticks(y_ticks)

# Utilização dos labels do plot para legendas
plt.legend()


#Exibicao e finalizacao do grafico
plt.show()

# ANÁLISES EXTRAS POR INICIATIVA PRÓPRIA

# Calculando a variação percentual de um ano para o outro
df_brasil['variacao_percentual'] = df_brasil['imigrantes'].pct_change() * 100

# Substituindo os valores NaN por 0 para evitar erro ao converter para inteiro
df_brasil['variacao_percentual'].fillna(0, inplace=True)

# Arredondando para remover casas decimais, convertendo para string com '%'
df_brasil['variacao_percentual'] = df_brasil['variacao_percentual'].round(0).astype(int).astype(str) + '%'

print(df_brasil)
```

```python
# ANÁLISES EXTRA DA ARGENTINA POR INICIATIVA PRÓPRIA

# Calculando a variação percentual de um ano para o outro
df_argentina['variacao_percentual'] = df_argentina['imigrantes'].pct_change() * 100

# Substituindo os valores NaN por 0 para evitar erro ao converter para inteiro
df_argentina['variacao_percentual'].fillna(0, inplace=True)

# Arredondando para remover casas decimais, convertendo para string com '%'
df_argentina['variacao_percentual'] = df_argentina['variacao_percentual'].round(0).astype(int).astype(str) + '%'

print(df_argentina)
```
